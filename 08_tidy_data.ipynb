```python
import numpy as np
import pandas as pd

# Each variable is a column.
# Each observation is a row.
# Each type of observation has its own separate data frame.

df = pd.read_csv("./git/bootcamp/data/gfmt_sleep.csv")
df["insomnia"] = df["sci"] <= 16
df.head()
```

```python
# Split - Apply - Combine => Aggregation
# Split
grouped = df.groupby(['insomnia', 'gender'])

# Apply & Combine (Median)
df_median = grouped.median(numeric_only=True)
# df_median

df_median = df_median.reset_index()
df_median
```

```python
# Transformation anstelle von Aggregation
df["rank grouped by insomnia"] = grouped["percent correct"].rank(
    method="first")

df.head()
```

```python
# SOrtierung der transformierten Werte
df_sorted = df.sort_values(by=["insomnia", "percent correct"])
df_sorted[["insomnia","percent correct","rank grouped by insomnia"]]
```

```python
import numpy as np
# Berechnung des Variationskoeffizienten
# V = S / M


def coeff(data=1):
    # print(data)
    # print(np.mean(data))
    result = np.std(data)/np.mean(data)
    return result

# TODO
# ValueError
grouped.agg(coeff)
```

```python
# Berechnung der Empirischen kumulativen Verteilungsfunktion

def ecdf_y(data):
    """Give y-values of an ECDF for an unsorted column in a data frame."""
    return data.rank(method='first') / len(data)

df['ecdf_y grouped by insomnia'] = grouped['percent correct'].transform(ecdf_y)
df.head()
```

```python
df.groupby(['gender', 'insomnia'])['percent correct'].describe()
```

```python
# Berechnung der Median für die Gruppierung Geschlecht & Schlaflosigkeit
for name, group in df.groupby(['gender', 'insomnia']):
    print(name, ': ', group['percent correct'].median())
```

```python
# Laden eines CSV-Files mit Daten
df = pd.read_csv('git/bootcamp/data/reeves_gradient_width_various_methods.csv', comment='#', header=[0,1])

df.head()

# Funktionsaufruf von melt mit den Variablen und dem Wert
df = pd.melt(df, value_name='gradient width', var_name=['genotype', 'method'])
# Entfernung der NaN-Werten
df = df.dropna()

df.head()
# Split Anhand
df.groupby(['genotype', 'method']).describe()
```
